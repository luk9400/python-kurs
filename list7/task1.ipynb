```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

tests = [
         {
            "title": "Parabola, 2 layers, sigmoid",
            "train": (np.linspace(-1, 1, 26), np.linspace(-1, 1, 26) ** 2),
            "model": tf.keras.Sequential([
               tf.keras.layers.Dense(10, activation="sigmoid"),
               tf.keras.layers.Dense(1)
            ]),
            "predict": np.linspace(-1, 1, 101)
         },
         {
            "title": "Parabola, 3 layers, sigmoid",
            "train": (np.linspace(-1, 1, 26), np.linspace(-1, 1, 26) ** 2),
            "model": tf.keras.Sequential([
               tf.keras.layers.Dense(10, activation="sigmoid"),
               tf.keras.layers.Dense(10, activation="sigmoid"),
               tf.keras.layers.Dense(1)
            ]),
            "predict": np.linspace(-1, 1, 101)
         },
         {
            "title": "Sine, 2 layers, tanh",
            "train": (np.linspace(0, 2, 21), np.sin((3 * np.pi / 2) * np.linspace(0, 2, 21))),
            "model": tf.keras.Sequential([
               tf.keras.layers.Dense(10, activation="tanh"),
               tf.keras.layers.Dense(1)
            ]),
            "predict": np.linspace(0, 2, 161)
         },
         {
            "title": "Sine, 3 layers, tanh",
            "train": (np.linspace(0, 2, 21), np.sin((3 * np.pi / 2) * np.linspace(0, 2, 21))),
            "model": tf.keras.Sequential([
               tf.keras.layers.Dense(10, activation="tanh"),
               tf.keras.layers.Dense(10, activation="tanh"),
               tf.keras.layers.Dense(1)
            ]),
            "predict": np.linspace(0, 2, 161)
         }
]

for test in tests:
  x, y = test["train"]

  plt.title(test["title"])
  plt.scatter(x, y)

  test["model"].compile(loss="mean_squared_error", optimizer=tf.keras.optimizers.Adam(0.1))
  test["model"].fit(x, y, epochs=1000, verbose=False)
  predicted = []
  for px in test["predict"]:
    predicted.append(test["model"].predict([px]))

  plt.scatter(test["predict"], predicted)
  plt.show()
```
